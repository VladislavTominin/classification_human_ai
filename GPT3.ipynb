```python
answers = []

with open(data_root/'train_set.json', 'r') as f:
    train_data = json.load(f)

def add_ghat_gpt_question(data):
    for sample in data:
        sample['gpt_question'] = 'Can you detect whether the following text was likely t have been generated by a large language model such as GPT3 or whether it was produced by a human? As an answer, indicate only the confidence score in percent. '
        sample['gpt_question'] += sample['text']
```

```python
add_ghat_gpt_question(train_data)
```

```python
for sample in train_data:
    chat = openai.ChatCompletion.create(
        model="gpt-3.5-turbo", messages=[ {"role": "user", "content": sample['gpt_question']}]
    )
    
    reply = chat.choices[0].message.content
    print(f"ChatGPT: {reply}")
    sample['gpt_answer'] = reply
```
